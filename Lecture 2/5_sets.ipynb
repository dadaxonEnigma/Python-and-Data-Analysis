## Sets 

Множества — одна из встроенных в Python структур данных. Одна из основных особенностей множеств заключается в том, что они не хранят повторяющиеся значения. Если вы попытаетесь добавить в множество повторяющееся значение, будет сохранено только одно из них.

Множества являются изменяемыми и неупорядоченными, то есть их элементы хранятся не в каком-то определённом порядке, поэтому для доступа к ним нельзя использовать индексы или ключи. Они могут содержать только значения неизменяемых типов данных, таких как числа, строки и кортежи. Они поддерживают математические операции над множествами, в том числе объединение, пересечение, вычитание и симметрическую разность.

```python
my_set = {1,2,3,4,5}
# my_set = {} # dict
# my_set = set() # empty set
```

```python
# Добавить элемент
my_set.add(5)
my_set.add(5)


# Удаление элемента
my_set.remove(5) # Ошибка если такого значения нет 
my_set.discard(5) # Ошибка не будет

# Очистка
my_set.clear()
```

## Методы issubset() и issuperset()

```python
my_set = {1, 2, 3, 4, 5}
your_set = {2, 3, 4, 6}

print(your_set.issubset(my_set))
print(my_set.issuperset(your_set))
```

## Метод isdisjoint()

```python
print(my_set.isdisjoint(your_set))
```

## Оператор | (объединение)
Возвращает новое множество содержащ все элементы из обоих множеств

```python
my_set | your_set
```

## Оператор & (пеересечение)

```python
my_set & your_set
```

## Оператор разности -

```python
my_set - your_set
```

## Оператор симметриной разности ^

```python
my_set ^ your_set
```

## Присваивание к переменной |= &= -= ^=

```python
my_set -= your_set

print(my_set)
```
