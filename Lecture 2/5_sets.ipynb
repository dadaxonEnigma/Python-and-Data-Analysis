## Sets 

Множества — одна из встроенных в Python структур данных. Одна из основных особенностей множеств заключается в том, что они не хранят повторяющиеся значения. Если вы попытаетесь добавить в множество повторяющееся значение, будет сохранено только одно из них.

Множества являются изменяемыми и неупорядоченными, то есть их элементы хранятся не в каком-то определённом порядке, поэтому для доступа к ним нельзя использовать индексы или ключи. Они могут содержать только значения неизменяемых типов данных, таких как числа, строки и кортежи. Они поддерживают математические операции над множествами, в том числе объединение, пересечение, вычитание и симметрическую разность.

```python
my_set = {1,2,3,4,5}
# my_set = {} # dict
# my_set = set() # empty set
my_second_set = set([1,2,3])
my_second_set
```

```python
# Добавить элемент
my_set.add(5)
my_set.add(5)


# Удаление элемента
my_set.remove(5) # Ошибка если такого значения нет 
my_set.discard(5) # Ошибка не будет

# Очистка
my_set.clear()
```

## Методы issubset() и issuperset()

```python
my_set = {1, 2, 3, 4, 5}
your_set = {2, 3, 4, 6}

print(your_set.issubset(my_set))
print(my_set.issuperset(your_set))
```

## Метод isdisjoint()

```python
print(my_set.isdisjoint(your_set))
```

```python
# Создайте функцию для поиска общих элементов между двумя списками
def find_common_elements(list1, list2):
    res = list(set(list1) & set(list2))
    print(sorted(res))

# Считываем два списка
list1 = ['apple', 'orange','banana','grape','melon']
list2 = ['orange','kiwi','melon','pear','grape']

# Находим и выводим общие элементы в алфавитном порядке
result = find_common_elements(list1, list2)
print(result)
```

```python
class BankAccount:
    # Ваш код тут
    def __init__(self, initial_balance):
        self.initial_balance = initial_balance
        
    def deposit(self, amount):
        if amount >= 0:
            self.initial_balance += amount
            return self.initial_balance
        else:
            return self.initial_balance
    
    def withdraw(self, amount):
        if amount <= self.initial_balance and amount >= 0:
            self.initial_balance -= amount
            return self.initial_balance
        else:
            return self.initial_balance
    
    def get_balance(self):
        return self.initial_balance
    
    

initial_balance = float(input())
deposit_amount = float(input())
withdraw_amount = float(input())

account = BankAccount(initial_balance)
print(account.get_balance())
print(account.deposit(deposit_amount))
print(account.withdraw(withdraw_amount))

```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Строковое представление
    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    # Перегрузка оператора сложения
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # Перегрузка оператора умножения (на скаляр)
    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

# Создаем векторы и используем перегруженные операторы
v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f"v1 = {v1}")  # вызывает __str__

print(f"v1 + v2 = {v1 + v2}")  # вызывает __add__

print(f"v1 * 2 = {v1 * 2}")  # вызывает __mul__

```
