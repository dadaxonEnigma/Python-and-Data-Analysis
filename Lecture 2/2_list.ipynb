## Метод append()

Вставляет элемент в конец списка. Если передать список как аргумент то он добваить его как влоенный список

```python
numbers = [1,2,3,4,5]
some_numbers = [6,7,8,9]

numbers.append(some_numbers)
print(numbers)
```

## метод extend()

Он работает как append но вместо вложенного списка он поэлемнтно добваляет в список

```python
numbers = [1,2,3,4,5]
some_numbers = [6,7,8,9]

numbers.extend(some_numbers)
print(numbers)
```

## List comprehensions

```python
even_numbers  = []

for num in range(21):
    if num % 2 == 0:
        even_numbers.append(num)

print(even_numbers)

# Но можно писать и по короче

even_numbers2 = [num for num in range(21) if num % 2 == 0]
print(even_numbers2)
```

```python
numbers = [1,2,3,4,5]
result = [(num, 'Even') if num % 2 == 0 else (num, 'Odd') for num in numbers]
print(result)
```

## Функция filter()

```python
words = ['tree', 'sky', 'mountain', 'river', 'cloud', 'sun']

def is_long_word(word):
    return len(word) > 4

long_words = list(filter(is_long_word, words))
print(long_words)
```

## Функция map()

```python
celsius = [0, 10, 20, 30, 40]

def to_fahrenheit(temp):
    return (temp * 9/5) + 32

fahrenheit = list(map(to_fahrenheit, celsius))
print(fahrenheit)
```

```python
def number_pattern(n):
    res = []
    if not isinstance(n, int):
        return 'Argument must be an integer value.'
    if n < 1:
        return 'Argument must be an integer greater than 0.'
    for num in range(1,n+1):
        res.append(str(num))
    
    return ' '.join(res)

print(number_pattern(4))
```
