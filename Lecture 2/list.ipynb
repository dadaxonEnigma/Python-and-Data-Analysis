```python
nums = [ 3,1,4,1,5,9]

# добавление
nums.append(2) # добовляет в конец
nums.insert(0,99) # добовляет по индексу
nums.extend([7,8]) # расширяет списком [..., 7,8]
```

```python
# Удаление 
nums.remove(1) # удоляет первое вхождения
popped = nums.pop() # удаляет и возвращает последний элемент
popped = nums.pop(2) # удаляет и возвращает элемент по индексу
```

```python
# Поиск и подсчёт
idx = nums.index(5) # индекс первого вхождения
cnt = nums.count(1)
cnt
```

```python
# порядок
nums.sort() # сортировка на месте (меняет список)
nums.sort(reverse=True) # по убыванию
nums.reverse() # разворачивает список
```

```python
# копия и очистка
copy = nums.copy() # поверхностное копия
nums.clear() # очищает список 
```

## Важное: sort() vs sorted()

```python
nums = [3,1,4]
nums.sort() # меняет nums, возвращает None
new = sorted(nums) # не меняет nums, возвращает новый список
```

## Ключ сортировки key =

```python
words = ["banana", "apple", "fig", "kiwi"]
words.sort(key = len) # по длине слова
words.sort(key=lambda w: w[-1]) # по соледней букве
```

Как сортировать по двум критериям сразу
key= принимает функцию, которая возвращает кортеж. Python сравнивает кортежи поэлементно — сначала первый элемент, потом второй:

```python
data = ["python", "java", "c", "javascript", "go", "rust", "kotlin"]

data.remove('go')
data.remove('c')
data.append('swift')
data.sort(key = lambda x: (len(x), x) )
data
```
