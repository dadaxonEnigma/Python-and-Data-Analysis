## Dictionary

Ключи в словаре должны быть уникальными и представлять собой неизменяемый тип данных.
Однако значения могут повторяться и иметь любой тип данных.

```python
pizza = {
    'name': 'Margharita Pizz',
    'price': 8.9,
    'calliries': 250,
    'toppings': ['mozzarella', 'basil']
}
pizza
```

## Метод pop() и popitem()
Метод .pop() удаляет пару «ключ-значение» с ключом, указанным в качестве первого аргумента, и возвращает его значение. Если ключ не существует, возвращается значение по умолчанию, указанное в качестве второго аргумента. Если ключ не существует и значение по умолчанию не указано, возникает исключение KeyError:


```python
pizza.pop('price',10)
```

```python
pizza = dict([('name', 'Margherita Pizza'), ('price', 8.9), ('calories_per_slice', 250), ('toppings', ['mozzarella', 'basil'])])

last_item = pizza.popitem()
print(last_item)
```

## Метод update()

обновляет пары «ключ-значение» с помощью пар «ключ-значение» из другого словаря. Если у них есть общие ключи, их значения перезаписываются.

```python
pizza.update({'price':15, 'total_time': 15})
pizza
```

## Обход словаря по циклу

```python
products = {
    'Laptop': 990,
    'Smartphone': 600,
    'Tablet': 250,
    'Headphones': 70,
}

for product, price in products.items():
    print( product ,price)
```

```python
products = {
    'Laptop': 990,
    'Smartphone': 600,
    'Tablet': 250,
    'Headphones': 70,
}

for product, price in products.items():
    products[product] = round(price * 0.8)

print(products)
```
