## Область видимости в Python

* Локальная (L): переменные, объявленные в функциях или классах
* Область видимости вкл функции (E): переменные, объявленные во включающих или вложенных функциях
* Глобальная область (G): переменные обявл на вкерхнем уррвне модуля или файла
* Встроенная область (B): зарезервированные в Python имена для предопределнныз функции, модлуей, ключевых слов и объектов

## Локальная область видимости

```python
def my_func():
    my_var = 10
    print(my_var)

my_func()
# print(my_var) # ошибка 
```

## Область видимости функции

Это озночает, что функция, вложенная в другую функцию, может обращаться к переменным функции, в которую она вложена

```python
def outer_func():
    msg = 'Hello there!'

    def inner_func():
        print(msg) # читаем переменную которая стоит на уровень выше

    inner_func()


outer_func()
```

!!! Но внешние функции не могут обращаться к переменным, определенным во вложенных функциях

```python
def outer_func():
    msg = 'Hello there!'
    # print(res) # res еще не инициализировано

    def inner_func():
        res = 'How are you?'
        print(msg)

    inner_func()
```

Одно из решений - инициализ res как пустую строку во внешней области видимости,
затем внутри inner_func сделать res нелокальной переменной с помощь ключевого слова nonlocal

```python
def outer_func():
    msg = 'Hello there!'
    res = ''
    print(res)

    def inner_func():
        nonlocal res 
        res = 'How are you?'

    inner_func()
    print(res)

outer_func()
```

## Глобальная облась видимости

Это переменные объявленные вне никаких-либо функции или классов, к которым мржно получить доступ из
любой точки программы.

```python
my_var = 100

def show_var():
    print(my_var)

show_var()
print(my_var)
```

Если хотите сделать локальную переменную, объявлениную внутри функции, доступной глобально, исп ключ слово global

```python
my_var_1 = 7
def show_vars():
    global my_var_2
    my_var_2 = 10
    print(my_var_1)
    print(my_var_2)

show_vars()

print(my_var_2)
```

Для изменения глоб переменной также можно исп global

```python
my_var = 10
def change_var():
    global my_var
    my_var = 20

change_var()

print(my_var)
```
