# Кортеж

Это упорядоченный последовательность, который могут содержать смешанные типы данных.
Кортежи явл неизменяемыми. После создание кортежа его элементы нельза изменить

```python
programming_languages = ('Python', 'Dart', 'Java', 'GO')
# programming_languages[0] = 'JavaScript'

# Доступ к элементу 
developer = ('Dadaxon', 19, 'ML Engineer')
developer[0], developer[-1]
```

## Другой способ создать кортеж - исп конструктор tuple()

```python
developer = 'Alisa'
tuple(developer)
```

## Проверка элемента 

```python
programming_languages = ('Python', 'Java', 'C++', 'Rust')

'Rust' in programming_languages # True
'JavaScript' in programming_languages # False
```

## Распаковка кортежа 

```python
developer = ('Alice', 34, 'Rust Developer')
name, age, job = developer
name, age, job
```

```python
developer = ('Alice', 34, 'Rust Developer')

name, *other = developer
name, other
```

## Методы работы с кортежами

## Метод count()

Показывает сколько раз элемент встречается в кортеже.

Если отсутствует элемент вернет 0
Если не передать аргумент вернет ошибку

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust')
programming_languages.count('Rust') # 2
```

## Метод index()

Поиск индекса элемента 

У него есть также атрибут позиии с которого нужно начать атрибут когда нужно закончить (в какой позиции)

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python')
print(programming_languages.index('Python')) # 2
print(programming_languages.index('Python', 4))
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python', 'JavaScript', 'Python')
programming_languages.index('Python', 2, 5) # 2
```

## Функция sorted()

Его также можно исп для кортежей

```python
programming_languages = ('Rust', 'Java', 'Python', 'C++', 'Rust', 'Python')
print('Key len',sorted(programming_languages, key=len))
print('Without Key',sorted(programming_languages))
print('With reverse',sorted(programming_languages, reverse=True))
```
