```python
def is_anogram(input_sting: str)-> bool:
    letters = []
    for ch in input_sting:
        if (ord(ch.lower())-96) < 1 or (ord(ch.lower())-96) > 26:
            continue
        if (ord(ch.lower())-96) in letters:
            continue
        else:
            letters.append(ord(ch.lower())-96)
    
    if sorted(letters) == list(range(1,27)):
        return True
    else:
        return False

print(is_anogram('The quick brown fox jumps over the lazy dog.'))
```

```python
def tower_builder(n_floors):
    tower = []
    max_width = n_floors * 2 - 1  

    for i in range(1, n_floors + 1):
        stars = '*' * (2 * i - 1)          
        floor = stars.center(max_width)    
        tower.append(floor)

    return tower
print(tower_builder(4))
```

```python
for row in tower_builder(4):
    print(row)
```
