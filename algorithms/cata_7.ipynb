## Задача 1 — Найти самую длинную строку
words = ["cat", "elephant", "dog", "lion", "tiger"]


```python
words = ["cat", "elephant", "dog", "lion", "tiger"]

def longest_word(words):
    longest = ''
    for w in words:
        if len(w) > len(longest):
            longest = w
    return longest
longest_word(words)
```

# Задача 5 — Удалить дубликаты, сохранив порядок
[1,2,2,3,1,4,2] → [1,2,3,4]

```python
def unique(lst):
    seen = set()
    res = []
    for x in lst:
        if x not in seen:
            seen.add(x)
            res.append(x)
    return res

print(unique([1,5,3,3,4,6,78,8]))
```

## Задача 6: преобразовать строку так, чтобы каждая буква стала её номером в алфавите.

```python
def alphabet_position(text):
    result = []
    for ch in text:
        if 'a' <= ch <= 'z':
            result.append(ord(ch.lower())-96)
    return result

print(alphabet_position('Hello, World!'))

# def alphabet_position(text):
#     return ' '.join(str(ord(ch.lower())-96) for ch in text if ch.isalpha())

# print(alphabet_position("The sunset sets at twelve o' clock."))
```

## Задача 10 - определить полендром

```python
def is_polindrom(s):
    s = s.lower().replace(' ','')
    return s == s[::-1]

# print(is_polindrom('An na'))

def is_polendrom(s):
    l = r = 0, len(s)-1
    s = s.lower()
    while l < r:
        if not s[l].isalnum(): 
            l+=1
            continue
        if not s[r].isalnum():
            r-=1
            continue
        if s[l] != s[r]:
            return False
        l += 1
        r -= 1
    return True
print(is_polendrom('hello'))
```

## Задача 19.


```python
def get_second_largest_number(nums: list[int]) -> int:
    first = second = float('-inf')

    for x in nums:
        if x > first:
            second, first = first, x
        elif first > x > second:
            second = x

    return second if second != float('-inf') else None

get_second_largest_number([3,42,1,413,2,3,1,56])
```

## Найти Максиму в списке

```python
def fin_max(arr):
    max_val = 0
    for x in arr:
        if x > max_val:
            max_val = x
    return max_val

print(fin_max([3,1,5,2]))
```

## Найти два числа, сумма которых равна target

```python
def two_sum(nums, target):
    d = {}
    for i,x in enumerate(nums):
        diff = target - x
        if diff in d:
            return (d[diff], i)
        d[x]=i

print(two_sum([2,7,11,15],9))
```

## Анаграмма

```python
def is_anagram(a, b):
    from collections import Counter
    return Counter(a) == Counter(b)

print(is_anagram("listen", "silent"))
```

## Максимальная длина подмассива 

```python
def max_window_sum(arr, k):
    curr = sum(arr[:k])
    best = curr
    for i in range(k, len(arr)):
        curr += arr[i] - arr[i-k]
        best = max(best, curr)
    return best

print(max_window_sum([1,5,2,3,7,1], 3))
```

## Найти второй максимум в списке

```python
def second_max(arr):
    first = second = float('-inf')
    for x in arr:
        if x > first:
            second = first
            first = x
        elif first > x > second:
            second = x
    return second

print(second_max([3,1,5,4]))
```

## Найти первое уникальное число

```python
def first_unique(arr):
    from collections import Counter
    cnt = Counter(arr)
    for x in arr:
        if cnt[x] == 1:
            return x

print(first_unique([4,4,2,1,2,5]))
```

## Bubble_sort

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

## SElection Sort

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```
