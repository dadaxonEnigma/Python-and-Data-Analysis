```python
arr = [10,20,10,30,5]

def max_subset(arr):
    k = 3
    window_sum = sum(arr[:k])
    max_sum = window_sum
    for i in range(k, len(arr)):
        window_sum += arr[i] - arr[i - k]
        if window_sum > max_sum:
            max_sum = window_sum
    return max_sum


print(max_subset(arr))
```

```python
def removeDuplicates(nums):
    if not nums:
        return 0
    
    k = 1
    
    for i in range(1, len(nums)):
        if nums[i] != nums[i - 1]:
            nums[k] = nums[i]
            k += 1
    
    return k

print(removeDuplicates([1,1,2,]))
```
