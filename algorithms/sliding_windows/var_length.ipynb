```python
def length_of_longest_substr(s):
    l = 0
    longest = 0
    sett = set()

    for r in range(len(s)):
        while s[r] in sett:
            sett.remove(s[l])
            l += 1
        sett.add(s[r])
        longest = max(longest,(r-l)+1)
    return longest

print(length_of_longest_substr('abcdeasdef'))
```
