## 217. Contains Duplicate



```python
# O(n) - 
# O(n) - 

def solution(input_nums: list[int]) -> bool:
    return False if len(set(input_nums)) == len(input_nums) else True
print(solution([1,2,3,1]))
```

```python
def solution(input_nums: list[int])->int:
    return sum(range(len(input_nums)+1)) - sum(input_nums)

print(solution([0,1,2,3,4,5,7]))
```

```python
def solution(input_nums):
    res = []
    set_nums = set(input_nums)
    for i in range(1,len(input_nums) + 1):
        if i not in set_nums:
            res.append(i)
    return res

print(solution([1,1]))
```
