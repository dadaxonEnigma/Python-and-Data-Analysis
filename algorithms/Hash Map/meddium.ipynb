```python
def groupAnagrams(strs: list[str]) -> list[list[str]]:
    groups = {}

    for words in strs:
        counter = [0] * 26

        for ch in words:
            index = ord(ch) - ord('a')
            counter[index] += 1
        key = tuple(counter)
        if key not in groups:
            groups[key] = []
        groups[key].append(words)
    return list(groups.values())
    

print(groupAnagrams(["act","pots","tops","cat","stop","hat"]))
```
