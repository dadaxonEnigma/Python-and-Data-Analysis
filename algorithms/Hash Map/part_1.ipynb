Массив не отсортирован поэтому исп **HashMap**.
Если массив был отсортирован то могли исп **Two Pointers**

```python
def twoSum(nums: list[int], target: int)-> list[int]:
    seen = {}
    for i, num in enumerate(nums):
        compliment = target - num
        if compliment in seen:
            return [seen[compliment], i]
        seen[num] = i
print(twoSum([2,7,11,15], target=9))
```

Время: O(n), Память: O(n)

# 2. Contains Duplicate

Дан массив nums. Верни True если есть дубликаты, False если все уникальны.
* [1,2,3,1] - True, 
* [1,2,3,4] - False

```python
nums = [1,2,3,1]
def contains_duplicate(nums):
    return True if len(nums) != len(set(nums)) else False
    # retrun len(nums) != len(set(nums)) короткий вариант
print(contains_duplicate(nums))
```

Время: O(n)
Память: O(n)

# 3. Valid Anagram

```python
s = 'anagram'
t = 'nagaram'

def valid_anagram(s, t):
    seen = {}
    if len(s) != len(t):
        return False
    for ch in s:
        seen[ch] = seen.get(ch, 0) + 1
    for ch in t:
        seen[ch] = seen.get(ch,0) - 1
    return all(v == 0 for v in seen.values())

print(valid_anagram(s,t))
```

```python
def find_max_profit(prices: list[int])->int:
    min_price = float('inf')
    max_profit = 0
    for price in prices:
        min_price = min(min_price, price)
        max_profit = max(max_profit, price - min_price)
    return max_profit

print(find_max_profit([2,4,1,7]))
```
