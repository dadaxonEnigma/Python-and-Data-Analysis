Массив не отсортирован поэтому исп **HashMap**.
Если массив был отсортирован то могли исп **Two Pointers**

```python
def twoSum(nums: int, target: int):
    seen = {}
    for i, num in enumerate(nums):
        compliment = target - num
        if compliment in seen:
            return [seen[compliment], i]
        seen[num] = i
print(twoSum([2,7,11,15], target=9))
```
