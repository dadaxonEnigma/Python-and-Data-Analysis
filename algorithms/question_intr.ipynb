## ЛОВУШКА СОБЕСЕДОВАНИЯ №3. default аргумент в функции

```python
def func(x, lst=[]):
    lst.append(x)
    return lst

print(func(1))
print(func(2))
print(func(3))
```

Потому что lst=[] создаётся ОДИН раз при объявлении функции
и сохраняется между вызовами.

```python
def cor_func(x, lst = None):
    if lst is None:
        lst = []
    lst.append(x)
    return lst

print(cor_func(1))
print(cor_func(2))
print(cor_func(3))
```

```python
def modify(lst):
    lst += [100]

x = [1, 2, 3]
modify(x)
modify(x)
modify(x)
print(x)
```

```python
def order_list(lst=None):
    pos = set()
    res = []
    for ch in lst:
        if ch not in pos:
            pos.add(ch)
            res.append(ch)
    return res

print(order_list([3,1,2,5,4,3,1]))
```

## Как работает прос-во имен
* LEGB (Local, Enclosingm Global, Built-in)

```python
str = 'global'

def outer():
    # str = 'enclosing'
    
    def inner():
        # str = 'local'
        print(str)
    inner()
outer()
```

## Как работают операторы global и nonlocal

```python
a = 1

def outer():
    b = 1
    
    def inner():
        global a
        a = 2
        nonlocal b
        b = 2

    inner()
    print("a =", a)
    print("b =", b)

outer()
```

## Функции map, filter, zip

```python
# map
a = [i for i in range(5)]

b = map(lambda x: x ** 2, a)

# print(list(b))

# filter
a = [i for i in range(5)]
b = filter(lambda x: x % 2 == 0, a)
# print(list(b))

# zip

a = [1,2,3]
b = [4,5,6,7]
c = [8,9,10]

for i in zip(a,b,c):
    print(i)
```

## Рекурсия

```python
def factorial(n):
    if n == 1:
        return n
    else:
        return n * factorial(n-1)
    
print(factorial(5))

# через цикл
def factorial(n):
    num = 1
    while n >= 1:
        num *= n
        n = n - 1
    return num
print(factorial(4))
```

## @classmethod и атрибут cls

## Как работают статический метод

```python
from datetime import datetime
class A:
    @staticmethod
    def get_current_datetime():
        return datetime.now()
    
print(A.get_current_datetime())
```
