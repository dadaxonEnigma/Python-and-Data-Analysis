## Two pointers 

Два указателя метод, где мы используем дваиндекса по одному массиву/строке и двигаем их по правилам, чтобы получить ответ за линейное (или близкое к линейному) время, вместо квадратичного перебора

### Когда использовать 
Используй, если хотя бы одна из этих характеристик истинна:
* Массив/строка отсортированы (или их можно отсортировать без потерь важной структуры)
* Ищещь пары/двойки элементов, удовлетворяющие условию (сумма, разность)
* Нуно сжать/удалить элементы in-place (удалить дубликаты, переместить нюли)
* Ищещь подстроки/подмассивы с каким-то свйоством

### Основыне вариации (и когда применять).


#### Пример 1 - Найти пару с суммой = target (отсортированный массив)

Задача: дан отсортированный массив nums, найти индексы двух чисел, сумма которых равна target/

```python
def two_sum(arr, target):
    left = 0
    right = len(arr) - 1

    while left < right:
        current_sum = arr[left] + arr[right]
        if current_sum == target:
            return [left, right]
        elif current_sum < target:
            left += 1 
        else:
            right -= 1

arr = [1, 3, 5, 7, 9]
print(two_sum(arr, 10))
```

#### Пример 2 - Remove duplicates from sorted array (in-place)

Задача: в отсортированном массиве удалить дубликаты так, что бы каждый элемент встречался толоько один раз; веруть новую длину

#### Пример 3 


```python
def squeares_sorted_array(nums: list[int])->list[int]:
    l, r = 0, len(nums) - 1
    res = []
    while l <= r:
        if abs(nums[l]) > abs(nums[r]):
            res.append(nums[l] ** 2)
            l += 1
        else: 
            res.append(nums[r]**2)
            r -= 1
    res.reverse()
    return res

print(squeares_sorted_array([-1,2,3,-5]))
```

```python
s = 'A man, a plan, a canal: Panama'

def check_polindrome(s: str)->bool:
    l, r = 0, len(s) - 1
    while l < r:
        while l < r and not s[l].isalnum():
            l += 1
        while l < r and not s[r].isalnum():
            r -= 1
        if s[l].lower() != s[r].lower():
            return False
        l += 1
        r -= 1
    return True

print(check_polindrome(s))
```

```python
s = 'leetcode'

def reverseVowels(s: str) -> str:
    # лучше исп set(vow) быстрее и убрать сравненеи lower() для каждой буквы 
    vow = set('aeiouAEIOU')
    list_s = list(s)
    l, r = 0, len(s) - 1
    while l < r:
        while l < r and list_s[l] not in vow:
            l += 1
        while l < r and list_s[r] not in vow:
            r -= 1
        list_s[l],list_s[r]=list_s[r],list_s[l]    
        l += 1
        r -= 1
    return ''.join(list_s)

print(reverseVowels(s))
```
