# 1. Boolean Expressions 
* True, False
* bool(), any(), all(), isinstance()

```python
print(True)
print(False)
print(type(True))
```

```python
print(bool(123))
print(bool(''))
print(bool(0))
print(bool(None))
```

## All, Any

```python
email = ""
phone = "193923023"
username = ""
print(all([email,phone,username]))
print(any([email,phone,username]))
```

## isinstance

```python
print(isinstance(123,int))
print(isinstance(True, str))

```

# 2. Comparison Operators


```python
print("a" < "b") # проверяет по алфовитном порядке
print("a" == "b") 
print("a" == "A")
```

```python
print(1 < 4 < 6) # проверяет  лево на право, все должно быть верной
```

# 3. Logical Operators
* and
* or
* not

```python
print( 3 > 2)
print( not 3 > 2 )
print(not False)
print(not not False) # вернет исходное положения
```

# 4. Control Mixed Conditions

```python
is_logged_in = True
is_guest = False
is_banned = True

print((is_logged_in or is_guest) and not is_banned)
```

# 5. Membership and Identity

* in
* not in
* is 
* is not

```python
print("o" in "python")
print("f" not in "python")

domain = 'gmail.com'
banned_domains = ["spam.com", "fake.org", "bot.net"]
print(domain not in banned_domains)
```

## is, is not 

Они проверяют ссылается ли два переменных в один и тотже объект в памяти

Пример 

a == b проверяет по значению
a is b проверяет по id (ссылку в памяти)

```python
a = 5
b = 5

a == b, a is b # для не сложных значений в памяти только один идентификатор
print(id(a))
print(id(b))
```

```python
x = ['a', 'b', 'c']
y = ['a', 'b', 'c']

print(x == y)
print(x is y)
print(id(x))
print(id(y))
```
