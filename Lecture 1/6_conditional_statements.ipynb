## Match case 
Исп только когда все совпадает. Нельза испл логические операторы

```python
country = 'Uzbekistan'

if country == 'United States':
    print('US')
elif country == 'Russian':
    print('RU')
elif country == 'Uzbekistan':
    print('UZB')
else:
    print('Unknown Country')

# ---------------------------------
# ---------------------------------
match country:
    case "United States":
        print('US')
    case 'Russian':
        print('RU')
    case 'Uzbekistan':
        print('UZB')
    case _:
        print('Unknown Country')
```
