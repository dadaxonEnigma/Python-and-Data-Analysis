* replace() -

```python
# Use case
# 1. Clean Numeric Formats
price = "$1,234.45"
print(price.replace(",","").replace("$",""))


# 2. Change phone number format
phone = "176-5434-56"
print(phone.replace("-","/"))

# 3. Remove elem with add empty ""
print(phone.replace('-',""))

phone = "+49 (176) 123-4567"
print('00'+ phone.replace("+","").replace("(","").replace(")","").replace("-","").replace(' ',""))
```

# Cleaning
* strip() - remove whitespace

```python
text = " Engineering   "
len(text) - len(text.strip())
len(text) == len(text.strip())

data = '968-Maria, ( D@t@ Engineer );; 27 y  '
data.lower().replace("@",'a').replace("(","").replace(')',"").replace(';','').strip()
```
