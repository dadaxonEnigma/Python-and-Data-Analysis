# Полиморфизм

Слово полиморфизм означает много форм. В програмирование оно относится к методам/функциям/операторам с одинаковыми именами, которые могут применяться ко многим объектам или классам.

Примером можно сказать функцию len()

```python
# Для string
x = 'Hello world!'
print(len(x))

# tuple
mytuple = ("apple", "bananana","grape")
print(len(mytuple))

# dict

thisdict = {
    "brand": "Ford",
    "model": "Focus",
    "year": 1964
}

print(len(thisdict))
```

## Полиморфизм классов

Он часто исп в метода классах, где у нас может быть несколько классов с одинаковыми методами.

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def move(self):
        print('Lets Go!')

class Ship:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def move(self):
        print('Swim')

class Plane:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):
        print("Fly")

car1 = Car('Ford', "Mustang")
boat1 = Ship('Ibiza', 'Touring 20')
palne1 = Plane('Boeing', "747")

for x in (car1, boat1, palne1):
    x.move()

# Благодоря полиморфизму мы може исп один и тот же метод для всех трех классов
```

## Полиморфизм классов наследования


```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def move(self):
        print('Movee')

class Car(Vehicle):
    pass

class Boat(Vehicle):
    def move(self):
        print('Swim Swim')

class Plane(Vehicle):
    def move(self):
        print('Fly !')

car1 = Car('For', 'Mustang')
boat1 = Boat('Ibiza', 'Touring 20')
plane1 = Plane('Boeing', '747')

for x in (car1, boat1, plane1):
    print(x.brand)
    print(x.model)
    x.move()
```

```python
class Cat:
    def sound(self):
        print('Meow')

class Fox:
    def sound(self):
        print('Wa-pa-pa-pa-pa-pow')

c1 = Cat()
f1 = Fox()

for x in (c1, f1):
    x.sound()
```

```python
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    
    def print_info(self):
        print(f'{self.brand} {self.year}')

car1 = Car('Tayota', 2020)
car1.print_info()
```

```python
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    
    def __str__(self):
        return (f'Car: {self.brand} ({self.year})')

car1 = Car('BMW', 2022)
print(car1)
```

```python
class Bird:
    def sound(self):
        print("Tweet")

class Dog:
    def sound(self):
        print('Woof')

def make_sound(animal):
    for a in animal:
        a.sound()

bird = Bird()
dog = Dog() 
make_sound((bird, dog))
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def sound(self):
        print(f'Woof, I am {self.name}')

    def __str__(self):
        return f'Woof, I am {self.name}'

class Cat(Animal):
    def sound(self):
        print(f'Meow, I am {self.name}')
    
    def __str__(self):
        return f'Meow, I am {self.name}'

first_dog = Dog('Rex')
first_cat = Cat('Luna')

for x in (first_dog, first_cat):
    x.sound()
print(first_cat)
print(first_dog)
```
