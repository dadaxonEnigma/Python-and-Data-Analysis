Типы методов в Python

* Обычные методы (методы экземпляров) - работают с конкретным объектом.
* Методы класса - работают с классом в целом.
* Статические методы - не работают ни с классом, ни с экземпляром
* Специальные методы - имеют специальное знаения в PYthon.

## Обычные методы (методы экземпляра)

Обычные методы - это функции, определенные внутри класса, которые принимают self в качестве первого параметра. Они могут обращаться к атрибутам экземпляра и вызывать другие методы экземпляра.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

rec = Rectangle(5, 3)
print(f'Площадь {rec.area()}')
print(f'Периметр {rec.perimeter()}')
```

## Методы Класса 

Это методы, которые принимают класс(а не экземпляр) в качетве первого параметра, обычно называемого cls. Они декорируются с помощью @classmethod и обычно исп для создания альтернативных конструкторов или для работы с атрибутами класса.

```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
    
    def display(self):
        return f'{self.day:02d}.{self.month:02d}.{self.year}'
    
    # Метод класса - альтернативный конструктор
    @classmethod
    def from_string(cls, date_string):
        day, month, year = map(int, date_string.split('.'))
        return cls(day, month, year)

# Создаем объект стандартным способом
date1 = Date(15,5, 2023)
print(date1.display())

# Создаем объект, используя метод класса
date2 = Date.from_string("19.08.2023")
print(date2.display())
```

## Статические методы

Это методы, которые не принимают ни self ни cls. Они декорируются с помощью @staticmethod и обычно исп для вспомогательных функции, логически связанных с классом, но не требующих доступа к состоянию класса или экземпляра.

```python
class MathUtils:
    @staticmethod
    def is_prime(number):
        """Проверяет является ли число простым"""
        if number < 2:
            return False
        for i in range(2, int(number**0.5) + 1):
            if number % i == 0:
                return False
        return True
        
# Вызываем статический метод через имя класса
print(f"Является ли 7 простым числом: {MathUtils.is_prime(7)}")
print(f"Является ли 10 простым числом: {MathUtils.is_prime(10)}")
```

## Специальные методы

Это методы с именами, начинающимися и заканчивающиеся двойными подчеркиваниеями. Они имеют специальное значение в Python и часто исп для перегрезки операторов или определения стандартного поведениям обхектов.

```python
class Vector:
    def __init__(self, x , y):
        self.x = x
        self.y = y
    
    # Строкове представление
    def __str__(self):
        return f'Vector({self.x}, {self.y})'
    
    # Перегрузка оператора сложения
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)
    
v1 = Vector(3,4)
v2 = Vector(1,2)
print(f'v1 = {v1}') # Вызывает __str__
print(f'v1 + v2 = {v1 + v2}') # Вызывает __add__
print(f'v1 * 2 = {v1 * 2}') # Вызывает __mul__
```
