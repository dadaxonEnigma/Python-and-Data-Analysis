## Класс и Объект
1. Класс определяет, как должен выглядеть объект
2. Объект создается на основе этого класса

Например:
* Фрукты (Класс) - Яблоко, Банан (Объект)

```python
class Person:
    '''Это докумнтация класса Person
    '''
    # Конструктор класса
    def __init__(self, name, age):
        self.name = name # атрибут экземпляра
        self.age = age # атрибут экземпляра
    
    # Метод класса
    def greet(self):
        return f'Привет Меня зовут {self.name}'

print(Person.__doc__)
```

## Изменения объектов после создания

```python
class Student:
    def __init__(self, name):
        self.name= name
        self.grades = []

    def add_grade(self, grade):
        self.grades.extend(grade)
        return f'Добавлена оценка: {grade}'
    
    def average_grades(self):
        if not self.grades:
            return 'Нет оценок'
        return f'Avarage: {sum(self.grades) / len(self.grades)}'
    
student = Student('Dadaxon')
print(f'Создан студент {student.name}')
print(student.average_grades())

student.add_grade([5,6,7,8])
student.average_grades()

# Добавляем новый атрибу (не рекомендуется но возможно)
student.age = 19
```

## __init__ method

Все классы имеют встроенный метод __init__(), который всегда выполняется при инициализации класса.
Метод __init__ исп для присвоения значений свойствам объекта или для выполнения операцией, при создании объекта


```python
# Метод __init__() вызывается автом каждый раз, когда класс исп для создания нового объекта
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Dadaxon", 24)

print(p1.name, end=' ')
print(p1.age)

# Без метода init() нам бы пришлось вручную задавайт свойстав для каждого объекта
class Person:
    pass

p1 = Person()
p1.name = 'Dadaxon'
p1.age = 25
```

## Параметр self

Параметр self - это ссылка а текущий экземпляр класса
Он исп для доступа к свойтсвам и метода, принадлежащим классу.

!!! self - должен быть первым параметром любого метода в классе

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        print(f"Hello. my name is {self.name}")
    
p1 = Person('Dadaxon',24)
p1.greet()
```

## Свойства класса в сравнении со совйствами объекта

Свойства определен внутри __init__() принадлежат каждому объекту

Совйства, определенные вне методов, пренадлежат самому классу и явл общими для всех объектов

```python
class Person:
    species = "Person" # Свойство класса

    def __init__(self, name):
        self.name = name

p1 = Person('Dadaxon')
p2 = Person('Ali')

print(p1.name)
print(p2.name)
print(p1.species)
print(p2.species)
```

## Несколько методов

В классе может быть несколько методов, которые работают вместе

```python
class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []
    
    def add_song(self, song):
        self.songs.append(song)
        print(f'Добавлено: {song}')

    def remove_song(self, song):
        if song in self.songs:
            self.songs.remove(song)
            print(f'Удалено: {song}')

    def show_songs(self):
        print(f'Плейлист "{self.name}"')
        for song in self.songs:
            print(f"- {song}")

my_playlist = Playlist("Favorite")
my_playlist.add_song("Bohemian Rhapsody")
my_playlist.add_song("Stairway to Heaven")
my_playlist.show_songs()
```

```python
class BankAccount:
    # Ваш код тут
    def __init__(self, initial_balance):
        self.initial_balance = initial_balance
        
    def deposit(self, amount):
        if amount >= 0:
            self.initial_balance += amount
            return self.initial_balance
    
    def withdraw(self, amount):
        if amount < self.initial_balance:
            self.initial_balance -= amount
            return self.initial_balance
        else:
            return f'Недостаточно денег на карте {self.initial_balance}'
    
    def get_balance(self):
        return f'Текущий баланс: {self.initial_balance}'
    
    

initial_balance = float(input())
deposit_amount = float(input())
withdraw_amount = float(input())

account = BankAccount(initial_balance)
print(account.get_balance())
print(account.deposit(deposit_amount))
print(account.withdraw(withdraw_amount))


```

## Когда использовать атрибуты класса vs атрибуты экземпляра

* Атрибуты экземпляра исп для данных, которые различаюься между объектами (имя, возраст, идентификатор)
* Атрибуты класса исп для:
    * Констант или настроек по умолчанию для всех экземпляров
    * Данных, которые должны быть общими для всех экземпляров
    * Отслеживания данных, связанных со всем классом (например кол-во созданнх экземпляров)
