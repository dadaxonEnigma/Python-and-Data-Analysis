Преимущество ООП

* Обеспечивает четкую структуру программ
* упрощает поддержку, повторное использование и откадку кода
* Позволяет создавать многократно использ приложения с меньшим кодом

## Класс и Объект
1. Класс определяет, как должен выглядеть объект
2. Объект создается на основе этого класса

Например:
* Фрукты (Класс) - Яблоко, Банан (Объект)

```python
class MyClass:
    x = 5

p1 = MyClass()
print(p1.x)
```

## Удаление объекта

```python
del p1
```

## Несколько объектов

!!! Каждый объект независим и имеет собственную копию свойств класса

```python
p1 = MyClass()
p2 = MyClass()
p3 = MyClass()

print(p1.x)
print(p2.x)
print(p3.x)
```

## __init__ method

Все классы имеют встроенный метод __init__(), который всегда выполняется при инициализации класса.
Метод __init__ исп для присвоения значений свойствам объекта или для выполнения операцией, при создании объекта


```python
# Метод __init__() вызывается автом каждый раз, когда класс исп для создания нового объекта
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Dadaxon", 24)

print(p1.name, end=' ')
print(p1.age)

# Без метода init() нам бы пришлось вручную задавайт свойстав для каждого объекта
class Person:
    pass

p1 = Person()
p1.name = 'Dadaxon'
p1.age = 25
```

## Значения по умолчанию в __init__()


```python
class Person:
    def __init__(self, name, age = 18):
        self.name = name
        self.age = age

p1 = Person("Dadaxon")
p2 = Person('Azik', 25)
print(p1.name, p1.age)
print(p2.name, p1.age)
```

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def bark(self):
        print(f"{self.name} says Woof!")

d1 = Dog("Buddy", 3)
d1.bark()
```
