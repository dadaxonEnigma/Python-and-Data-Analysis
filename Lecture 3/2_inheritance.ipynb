# Наследование Python

Наследование позволяет определить класс, который наслдеуеь все методы и свойств другого класса
* Родительский класс - это класс, от которого происходит наследование
* Дочерний класс - это класс, который наследует от другого класса

```python
class Person:
    def __init__(self, fname = 'Dadaxon', lname="Unknown"):
        self.firstname = fname
        self.lastname = lname
    
    def printname(self):
        print(self.firstname, self.lastname)

x = Person('Dadaxon', 'Turgunboev')
x.printname()
```

## Дочерний класс

```python
class Student(Person):
    pass # если не хотите добавить свой или методы 

first_student = Student(lname='Olsen')
first_student.printname()
```

!!! Если добавить функ __init__() дочерний класс больше не будет гаследовать функцию __init__() родительского класса

```python
class Student(Person):
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname
    
second_student = Student('','')
second_student.printname()
```

!!! Что бы сохранить наследоваие родит класса 

```python
class Student(Person):
    def __init__(self, fname, lname):
        Person.__init__(self, fname, lname)
```

## Функция super()

Он позволяет дочернему классу наследовать все методы и свойства родительского класса

```python
class Student(Person):
    def __init__(self, fname, lname):
        super().__init__(fname, lname)
```

## Добваления свойств

```python
class Student(Person):
    def __init__(self, fname, lname):
        super().__init__(fname, lname)
        self.graduationyear = 2026

x = Student("Deny", "Olise")
x.graduationyear
```

```python
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        print(f'{self.name}')

class Dog(Animal):
    pass

d1 = Dog('Rex')
d1.speak()
```
