## Инкапсуляция

Инкапсуляция - то защита данных внутри класса.

Это озночает что данные(свойсвта) и метрды хранятся в одном лксаае, а доступ к данным изне контролируется

Это предотврощает случайное изменение данных и скрывает внутренние детали работы класса

```python
class Wallet:
    def __init__(self, balance):
        self.__balance = balance # можно исп только внутри класса

    def deposit(self, amount):
        self.__balance += amount
    
    def withdraw(self, amount):
        self.__balance -= amount

    def get_balance(self):
        return f'Баланс: {self.__balance}'

acc1 = Wallet(1000)
acc1.deposit(1000)
acc1.get_balance()
acc1.withdraw(500)
acc1.get_balance()
```

## Мы также можем сделать привытный метод

```python
class Wallet:
   def __init__(self):
       self.__balance = 0

   def __validate(self, amount):
       if amount < 0:
           raise ValueError('Amount must be positive')

   def deposit(self, amount):
       self.__validate(amount)
       self.__balance += amount

   def withdraw(self, amount):
       self.__validate(amount)
       if amount > self.__balance:
           raise ValueError('Insufficient funds')
       self.__balance -= amount

   def get_balance(self):
       return self.__balance

acct_one = Wallet()
acct_one.deposit(3)
print(acct_one.get_balance()) # 3

acct_one.deposit(50)
print(acct_one.get_balance()) # 53

# acct_one.deposit(-4)  # ValueError: Amount must be positive
# acct_one.withdraw(-8) # ValueError: Amount must be positive
# acct_one.withdraw(58) # ValueError: Insufficient funds
```

## Получить значение приватного свойства

Нужно создать метод

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age
    
    def get_age(self):
        return self.__age
    
p1 = Person("dadaxon", 25)
p1.get_age()
```

Значение частного свойства

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age
    
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print('Возраст должен быть положительным')

p1 = Person('Ali', 24)
print(p1.get_age())

p1.set_age(30)
print(p1.get_age())
```

## Зачем исп Инкапсуляцию
* Защита данных: предотврощает случайное изменение данных
* Проверка: Вы можете проверить данные перед их установкой
* Контроль: Вы полностью контролируете доступ к данными и их изменением

Пример

Используйте инкопсуляцию для защиты и проверки данных

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.__grade = 0

    def set_grade(self, grade):
        if 0 <= grade <= 100:
            self.__grade = grade
        else:
            print('Оценка должна быть в диапозоно от 0 до 100')
    
    def get_grade(self):
        return self.__grade

    def get_status(self):
        if self.__grade >= 60:
            return 'Pass'
        else:
            return 'Fail'
        
s1 = Student('Dany')
s1.set_grade(80)

print(s1.get_grade())
print(s1.get_status())
```

## Защищенные свойства

В Python также принято исп префик _ для обозначения защищенных свойств:

!!! Одинарное подчеркивание - жто просто условность. Оно сообщает другим програмистам что свойсвто предназначено для внутренного исп

```python
class Person:
    def __init__(self, name, salary):
        self.name = name
        self._salary = salary # защищенное свойство

p1 = Person('Anna',50000)
print(p1.name)
print(p1._salary) # можно получить доступ но не следует этого делать
```

## Приватные методы

```python
class Calculator:
    def __init__(self):
        self.__result = 0
    
    def __validate(self, num):
        if not isinstance(num, (int, float)):
            return False
        return True
    
    def add(self, num):
        if self.__validate(num):
            self.__result += num
        else:
            print('неверное число')

    def get_result(self):
        return self.__result

calc = Calculator()
calc.add(10)
calc.add(5)

print(calc.get_result())
# print(calc.__validate(4)) # Это приведет к ошибке
```
