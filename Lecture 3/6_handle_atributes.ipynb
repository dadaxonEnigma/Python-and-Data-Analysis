## Как динамически управлять атрибутами объекта

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

my_car = Car('Chevrolet','Onix')
print(my_car.brand)
print(my_car.model)
```

Но иногда вы не можете знать, какаие атрибуты вам нужны, пока программа не запущена. 
Таким обрахом, вы можете получать доступ к атрибутам, изменять их, проверять или даже удолять

В Python есть 4 встроенные функции для динамической работы с атрибутами объекта.

* getattr()
* setattr()
* hasattr()
* delattr()

## getattr()

Позволяет считать атрибут объекта, если его назвавние неизвестно до выполнения программы.
Если атрибут не сущесвует, возникает исключение AttributeError если только вы не укажете значение по умолчанию

```python
class Person: 
    def __init__(self, name, age): 
        self.name = name 
        self.age = age 

person = Person('John Doe', 30) 

print(getattr(person, 'name'))
print(getattr(person, 'age'))
print(getattr(person, 'city', 'Milano'))
```

## dir()

Просмотр всех атрибутов объекта

```python
class Person: 
    def __init__(self, name, age): 
        self.name = name 
        self.age = age 

person = Person('John Doe', 30)

for attr in dir(person):
    if not attr.startswith('__') and not callable(getattr(person, attr)):
        value = getattr(person, attr)
        print(f'{attr}: {value}')
    
```

## setattr()

Позволяет динамически создавать новый атрибут или обнавлять существующий

```python
class Configuration:
    pass

settings_data = {
    'server_url': 'https://api.example.com',
    'timeout_sec': 30,
    'max_retries': 5
}

config_obj = Configuration()

for attr_name, attr_value in settings_data.items():
    setattr(config_obj, attr_name, attr_value)

print(config_obj.__dir__())
print(config_obj.timeout_sec)
```

## hasattr()

Он прловеряет сущесвтует ли атрибут и в зависимости от результата возвращает True или False

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

product_a = Product('T-Shirt', 25)

required_atrributes = ['name', 'price', 'inventory_id']

for attr in required_atrributes:
    if not hasattr(product_a, attr):
        print(f"ERROR: Product is missing the required attribute: '{attr}'")
    else:
        print(f'{attr}: {getattr(product_a, attr)}')
```

## delattr()

Позвояет динамически удолять атрибут

```python
class UserSession:
    def __init__(self, user_id, token):
        self.user_id = user_id
        self.auth_token = token # sensitive
        self.temp_counter = 0 # temporary

session = UserSession(101, 'a1b2c3d4e5')

attributes_to_clean = ['auth_token', 'temp_counter']

for attr in attributes_to_clean:
    if hasattr(session, attr):
        delattr(session, attr)
        print(f'removed: {attr}')

print('\nFinal attribute remaining:')

for attr in dir(session):
    if not attr.startswith('__') and not callable(getattr(session, attr)):
        print(f'- {attr}: {getattr(session, attr)}')
```
