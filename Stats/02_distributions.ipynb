# Распределение данных и распределение выборок

* Выборка (sample)
* Популяция (population)
* Случайный отбор (random sampling)
* Стратифицированный отбор
* Смещенная выборка (sample bias)

**Выборка** - это подмножество данных из более крупного набора данных(популяция или ген совакупность).

**Случайный отбор** - это процесс, вкотором каждый член популяции, подвергаемой отбору, имеет равную возможность попасть в выборку при каждом извлечении

**Стратифицированный отбор** - разделения популяции на страты (подгруппы) и случайный отбор из каждой страты.

**Смещенная выборка** - Выборка, которая представляет популяцию в искаженном виде

## Выборочная распределение статистической велечины

* Центральная предельная теорема
* Стандартная ошибка 

**Ценральная предельная теорема** - тенденция выборочного распределения принимать нормальную форму по мере увелечени размера выборок

**Стандартная ошибка** - это одиноная метрика, которая обобщает вариабельноть статистической велечины в выборочном распределении.

## Бутстрап и повторный отбор

**Бутсраповская выборка** - выборка, взятая с возвратом из набора данных
**Повторный отбор(resampling)** - процесс многокртаного взятия выборок из наблюдаемых данных

 Бутстрап не компенсирует малый размер выборки; он не создает новые данные 
и при этом не заполняет дыры в существующем наборе данных. Он просто 
сообщает о том, как поведут себя многочисленные дополнительные выборки, 
когда они будут извлекаться из популяции, такой как наша исходная выборка.
