```python
import re
```

```python
book = 'Fahrenheit 451'

print(re.search('\d', book))
print(re.search('\d+', book))
```

```python
print(re.fullmatch("\d+", book))
print(re.fullmatch('Fahrenheit \d+',book))
```

```python
numbers = [10,20,30,40]

# Добавьте число 100 в конец списка
numbers.append(100)

# Удалите первый элемент из списка
del numbers[0]

print(' '.join(map(str, numbers)))
# Выведите измененный список в одну строку, с элементами, разделенными пробелами
```
