```python
def check_age(age):
    if age < 0:
        raise ValueError('Age cannot be negative')
    return age

try:
    check_age(-5)
except ValueError as e:
    print(e)
```

```python
def process_data(data):
    try:
        result = int(data)
        return result * 2
    except ValueError:
        print('Logging: Invalid data received')
        raise

try:
    process_data('ad')
except ValueError as e:
    print(e)
```

```python
def increment_string(s):
    i = len(s) - 1
    
    # идём с конца, пока встречаем цифры
    while i >= 0 and s[i].isdigit():
        i -= 1
    
    # всё, что справа — это число
    prefix = s[:i+1]
    number = s[i+1:]
    
    if number:
        new_number = str(int(number) + 1).zfill(len(number))
        return prefix + new_number
    else:
        return s + "1"


        
print(increment_string('foo041'))
```
